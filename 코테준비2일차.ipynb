```python
#실전문제 2. 개임개발
#방향개념이 도입되어서 풀기가 쉽지 않음
#모범답안 풀이로 문제를 이해하려 시도
n, m = map(int, input().split())
x, y, direction = map(int, input().split())

map_ = []
for i in range(n): #줄마다 바뀌는 값들을 받는 방법을 익히기
    map_.append(list(map(int, input().split())))
    
#왼쪽 방향으로 회전하고 왼쪽으로 한칸 전진한다는 의미를 책과 내가 다르게 이해하는 것 같음
#책의 풀이는 왼쪽으로 회전하고 왼쪽으로 이동을 왼쪽으로 회전한 상태에서 직진형태로 구현하고 있음
#본인은 왼쪽 회전이후 왼쪽방향으로 전진하는것으로 이해함

dx = [-1, 0, 1, 0] #북 동 남 서 방향으로 이동할 경우의 좌표 변화 입력
dy = [0, 1, 0, -1]

#왼쪽 회전에 대한 함수 정의
def turn_left():
    global direction #direction이라는 전역변수에 변화를 주기위해서 이와 같이 입력함
    direction -= 1
    if direction == -1:
        direction = 3

#방문한 곳을 파악하기 위해 빈 행렬을 생성
empty = [[0] * m for _ in range(n)] 
    #[][]를 인덱스로 사용하여 접근함. 이 때 인덱스로 사용될 수 있는 것은 x,y의 위치 
        
#시작
count = 1 #방문한 곳. 현재 있는 위치도 이미 방문한 것으로 처리
turn_time = 0 #회전한 횟수

while True:
    
    #회전하고 먼저 이동해봄. 계속해서 반복할 행동
    turn_left()
    nx = x + dx[direction]
    ny = y + dy[direction]
    
    #이동한 곳이 처음 간 곳일경우
    if empty[nx][ny] == 0 and map_[nx][ny] == 0: #실제 육지인 경우도 포함하여 검증
        empty[nx][ny] = 1 #간 곳을 기록하기
        x = nx
        y = ny
        count += 1
        turn_time = 0 #다시 처음부터 돌아가야 하므로 회전횟수를 0으로 초기화
        continue #회전 후 이동이 이루어졌다면 아래의 조건들을 수행할 필요가 없으므로 무시하는 continue
        
    #이동한 곳이 이미 갔던 곳이거나 바다일 때
    else:
        turn_time += 1 
        #돈 횟수만 추가한다. 
        #만약 이 코드 아래에 해당사항이 없을때는 다시 처음부터 돌아와서 돌고 이동하는 작업을 수행한다
        
    if turn_time == 4: #4방향으로 모두 회전한 경우. 즉 갈 곳이 없을 때 뒤로가기
        nx = x - dx[direction]
        ny = y - dy[direction]
        
        if map_[nx][ny] == 0: #실제 지도에서 뒤로간 곳이 육지에 해당할 때
            x = nx
            y = ny
        
        else: #뒤로갔는데 바다일 경우에
            break
        turn_time = 0 #회전을 초기화하고 반복을 다시 수행해야함
    
print(count)
```

## While 문 작성 요령
- While True: 바로 밑에는 **계속해서 반복될 코드**를 작성한다
- 반복되는 해당 코드의 결과를 확인할 수 있는 객체도 선언한다
- 해당 객체의 결과를 if, else구문을 통해서 수시로 조건을 확인한다
- 만약 특정한 조건에 해당된다면 그에 따른 추가 코드를 if, else 밑에 작성한다
- 중요한 것은 While바로 밑에 반복되는 코드는 조건이 맞는 한 계속 반복되어야 하도록 작동해야 함

```python
#재귀함수 2
#팩토리얼

def factorial_recursive(n):
    if n <= 1: #0!, 1!에 대한 예외처리
        return 1
    return n * factorial_recursive(n-1)

print('재귀적으로 구현:', factorial_recursive(4))
```

```python
#재귀함수 3
#유클리드 호제법
#두 자연수에 대한 최대공약수를 구할때 사용되는 대표적인 알고리즘
#A > B 일때 A와 B의 최대공약수는 A와 B를 나눈 나머지 R과 B의 최대공약수와 같다
#즉 A % B = R, B % R = C, .... 형식으로 나머지가 없을 때 까지 나눗셈을 반복하는 형식으로 함수를 작성하면 됨 

def gcd(a,b):
    if a % b ==0:
        return b
    else:
        return gcd(b, a % b)
    
print(gcd(192, 162))
```

### DFS, Depth-First Search
- 깊이 우선 탐색, 말 그대로 그래프에서 깊은 부분을 우선적으로 탐색하는 알고리즘
- **스택 자료구조 (재귀함수)**를 이용해서 구현하며 동작 과정은 다음과 같음
    - 탐색 시작 노드(첫 시작 노드)를 스택에 삽입하고 방문 처리함
    - 스택 최상단 노드(**가장 나중에 들어온** 노드)에 방문하지 않은 인접 노드가 하나라도 있다면 -> 해당 인접 노드를 스택에 넣고 방문처리함
    - 만약 없다면 -> 최상단 노드를 꺼냄
    - 위의 과정을 반복수행할 수 없을 때 알고리즘을 멈춤
    
- 문제에서는 방문 과정을 주게 된다 (예로 낮은 번호부터 방문)

```python
#DFS 예제
#그래프에서 DFS를 노드를 탐색하는 순서를 출력

graph = [ #각 노드별 다른 노드와 연결된 관계를 리스트 형태로 표현
    [],
    [2, 3, 8], #1번째 노드와 연결된 다른 노드들의 번호
    [1, 7],
    [1, 4, 5],
    [3, 5],
    [3, 4],
    [7],
    [2, 6, 8],
    [1, 7]
]

visited = [False] * len(graph) #방문 여부를 표현하는 객체. False, True 형태 Boolean으로 표현

def dfs(graph, v, visited): 
    #graph는 노드들의 연결관계 표현 객체, v는 입력 노드 번호, visited는 노드별 방문 여부를 나타내는 객체
    
    visited[v] = True #현재 입력된 노드 번호를 방문 처리
    print(v, end = ' ')
    for i in graph[v]:
        if not visited[i]: #인접 노드가 방문되지 않은 경우에만 재귀적 실행
            dfs(graph, i, visited)
            
dfs(graph, 1, visited)
```

### BFS, Breath First Search
- 너비 우선 탐색, 가까운 노드부터 우선적으로 탐색하는 알고리즘
- **큐 자료구조**를 이용
    - 탐색 시작 노드(첫 노드)를 큐에 삽입하고 방문처리
    - 큐에서 노드를 꺼낸 뒤, 해당 노드의 인접 노드 중 방문하지 않은 모든 노드를 큐에 삽입하고 방문 처리
    - 위의 사항을 반복하지 못할때까지 처리
- DFS와 마찬가지로 방문과정은 문제에서 주어짐 (낮은 번호순부터)

```python
#BFS 예제
graph = [ #각 노드별 다른 노드와 연결된 관계를 리스트 형태로 표현
    [],
    [2, 3, 8], #1번째 노드와 연결된 다른 노드들의 번호
    [1, 7],
    [1, 4, 5],
    [3, 5],
    [3, 4],
    [7],
    [2, 6, 8],
    [1, 7]
]

visited = [False] * len(graph)

from collections import deque

def bfs(graph, start, visited):
    
    queue = deque([start]) #시작 노드를 큐에 넣기
    
    visited[start] = True #시작 노드를 방문 처리
    
    while queue: #큐가 비워지지 않는한 반복
        v = queue.popleft() #큐의 최상단 노드부터 출력. 즉 먼저 들어온 순서대로 출력하게 됨
        print(v, end = ' ')
        for i in graph[v]:
            if not visited[i]:
                queue.append(i) #인접 노드들을 순서대로 큐에 삽입하기
                visited[i] = True #삽입된 노들들을 방문처리함
        
bfs(graph, 1, visited)
```

```python
#실전문제 1. 음료수 얼려먹기
#DFS, BFS 개념이 명확하지 않으므로 풀이를 보면서 문제를 풀기
#0이 뚫린경우, 1 이 막힌 경우. 상하좌우로 붙어있는경우는 연결되어 있다고 가정. 
#연결되어 있을 때는 하나의 틀로 사용될 수 있음
#아이스크림을 생성할 수 있는 연결체들의 최대개수를 구하는것과 동일
#각 0, 노드별 최대 연결되어 있는 개수 -> 하나의 연결체로 count -> DFS

n,m = map(int, input().split())
graph = []
for i in range(n):
    graph.append(list(map(int, input())))
    
def dfs(x, y):
    if x <= -1 or x >= n or y<= -1 or y >= m:
        return False #x, y가 가지는 범위가 n,m 혹은 -1 이하에 해당될 때 종료
    if graph[x][y] == 0: #첫 노드, 첫 노드로부터 뻗은 인접노드가 빈 곳에 해당하는 경우
        graph[x][y] = 1 #해당 노드의 방문 처리. 즉 음료수를 붓는 행위
        dfs(x-1, y) #상, 하, 좌, 우로 가서 조건이 맞으면 붙는 행위
        dfs(x, y -1)
        dfs(x+1, y)
        dfs(x, y + 1)
        return True #방문이 끝나면 하나의 묶음으로 처리하기위해 True를 반환
    return False #애초에 시작이 1에 해당한다면 탐색을 진행하지 않음

result = 0 #아이스크림 갯수

for i in range(n):
    for j in range(m):
        if dfs(i,j) == True:
            result += 1
            
print(result)
```

## 본 문제를 통해 알아본 DFS형 문제풀이
- 어떤 깊이를 최대한 탐색하는 것은 곧 최대한 연결할 수 있는 하나의 연결체를 구성하는 것과 동일하다
- 연결체를 구성할 수 있는 함수를 dfs로 명명하고 재귀적으로 판단하는 지점 및 재귀를 종료하는 조건을 입력해야 한다
- 하나의 연결체가 구성된다면 단순히 0, 1 혹은 True, False형태의 출력을 내면 된다

```python
#미로 탈출
#0은 괴물, 1은 통로. 0을 만나지 않아야 함
#시작칸, 끝칸을 포함하여 이동할 최소한의 칸의 개수
#최소칸의 개수를 찾으려면 건널 칸의 순서가 중요. -> 큐 자료구조에서 가장 먼저 나가는 노드의 개수를 바탕으로
#나간 노드의 인덱스를 바탕으로 끝점에 도달하면 중단하기
from collections import deque

n,m = map(int, input().split())
graph = []
for i in range(n):
    graph.append(list(map(int, input())))

#상하좌우로 이동하는 기법을 사용
dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]
    
def bfs(x,y):
    queue = deque()
    queue.append((x, y))
    
    while queue:
        x,y = queue.popleft()
        
        for i in range(4):
            nx = x + dx[i] #상, 하, 좌, 우로 이동이 가능한지 시도
            ny = y + dy[i]
            
            if nx < 0 or nx >= n or ny < 0 or ny >= m:
                continue #가능한 인덱스 위치를 벗어나는 경우 아래의 경우를 무시
                
            if graph[nx][ny] == 0: #괴물이 있는 경우 무시
                continue
                
            if graph[nx][ny] == 1:
                graph[nx][ny] = graph[x][y] + 1 #시작 지점의 값에서 1씩 추가. 즉 이동한 칸의 개수를 세는 것과 마찬가지
                queue.append((nx, ny)) #새로 시작할 인덱스를 큐에 삽입
    return graph[n-1][m-1] #가장 마지막 위치까지 이동한 칸의개수를 출력

print(bfs(0, 0))
```

## 본 문제를 통해 알아본 BFS형 문제풀이
- 그래프 형태의 경로의 최단 이동수를 알기 위해서는 BFS가 유용하다
- BFS를 이용하여 다른 인접노드를 탐색하고자 할 때 dx,dy같은 방향벡터를 사용하여 이동을 시도하고 검증하는 것이 유용하다
- 이동이 가능한 인접행렬을 모두 이동가능한 큐에 넣고 그 중에서 다시 그 노드의 인접행렬로 이동하고자 하기 때문에 최단거리를 찾을 수 있다
- 이동한 노드에 대하여 문제에서 요구하는 형식으로 노드번호를 변경할 수 있다 (예로 노드의 값을 이동한 칸의 개수로 대체)
